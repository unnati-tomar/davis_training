```
num_str = input("Enter a number: ")

greatest_digit = 0
for digit_char in num_str:
    digit = int(digit_char)
    if digit > greatest_digit:
        greatest_digit = digit

print(f"greatest digit -- {greatest_digit}")
for x in range (10):
      print (x)
```

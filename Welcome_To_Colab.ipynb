```

#Find out greatest digit present in given number
num_str = input("Enter a number: ")

greatest_digit = 0
for digit_char in num_str:
    digit = int(digit_char)
    if digit > greatest_digit:
        greatest_digit = digit

print(f"greatest digit -- {greatest_digit}")
for x in range (10):
      print (x)
```

```
#to find out factorial of a number
num = int(input("Enter any number: "))

if (num < 0):
  print("Factorial not possible")
elif(num == 0):
   print("Factorial = 1")
else:
  fact = 1
  for x in range (1, num + 1):
    fact = fact * x
  print("Factorial = ", fact)

```

```
#to display even factor of a number
num = int(input("Enter a number: "))

print(f"Even factors of {num} are:")
for i in range(1, num + 1):
    if num % i == 0 and i % 2 == 0:
        print(i)

```

```
#to find hcf of 2 numbers
num1 = int(input("Enter first number:"))
num2 = int(input("Enter second number:"))
if(num2 == 0):
  print("HCF :",num1)
else:
    while(num2 != 0):
      rem = num1 % num2
      num1 = num2
      num2 = rem
      print("HCF :",num1)
```
